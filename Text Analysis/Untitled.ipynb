```python
result = pd.merge(rest,df,left_on = 'resId ',
                   right_on = 'resId',how = 'inner')
```

```python
result = result.drop('resStreet',1)
```

```python
result.head()
```

```python
result1 = result[['resId ','resName','resCity','resState','resStars','resRevCNT','resPrice','resTransactionType','vote','resLat','resLong']]
```

```python
result1.head()
```

```python
result1.to_excel('finalanalysis1.xlsx')
```
